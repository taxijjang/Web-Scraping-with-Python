# 크롤링 정리
1. 크롤링할 html 코드를 가져온다. (requests.get())
2. 파싱한다. (BeautifulSoup(res.content, 'html-parser'
3. 특정 데이터를 선택한다.
4. 출력한다. (get_text())

```python
from bs4 import BeautifulSoup
import requests

res = requests.get("https://sports.news.naver.com/index.nhn")
html = res.content
soup = BeautifulSoup(html, 'html.parser')

title = soup.find_all('span',{'class' : 'title'})

for d in title:
    print(d.get_text())
```
